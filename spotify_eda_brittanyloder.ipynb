```python
#drop "Unnamed" column (it's probably an index brought over from the original csv)
df.columns
df = df.drop('Unnamed: 0', axis=1)
```

```python
#Central Tendency: the location of middle values (mean, median, mode)
#Spread: how far from the middle the values in a distribution fall (variance, standard deviation, interquartile range)
#Skewness: measure of symmetry in data
#Kurtosis: measure of how much of the data falls on the tails/edges of the distribution
df.describe()
```

```python
df.columns
df_corr = df[['popularity',
       'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr
```
